## Box and Violin Plots

**Box plot** can show the range of data

**Violin plots** are an enhancement of boxplot.  In addition to the range of data, it also shows the distribution of data

References

- https://www.python-graph-gallery.com/boxplot/
- https://www.python-graph-gallery.com/violin-plot/

## 1 - Load Data

```python
import pandas as pd
import seaborn as sns
```

```python
df = sns.load_dataset('tips')
df.sample(10)
```

## 2 - Boxplot

Let's start with a familiar boxplot.

**Q=> What can you infer from the boxplot?**

```python
sns.boxplot(data=df, x='day', y='tip')
```

```python
sns.boxplot(data=df, x='time', y='tip')
```

## 3 - Add Comparisons

```python
# compoare lunch vs dinner

sns.boxplot(data=df, x='day', y='tip', hue='time')
```

```python
## Do smokers tip more?

sns.boxplot(data=df, x='time', y='tip', hue='smoker')
```

## 4 - Violin Plot

Interpretting the results:

- the tip ranges from 0 to 11 (?)
- but the width of the graph shows the distribution.  So we have more in the 2-4 range

```python
sns.violinplot (data=df, y='tip')
```

## 5 - Let's compare

Let's do a day by day comparison

```python
sns.violinplot (data=df, x='day', y='tip')
```

```python
# how about lunch/dinner
sns.violinplot (data=df, x='time', y='tip')
```
